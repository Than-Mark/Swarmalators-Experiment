$$
\begin{cases}
	\dot{\mathbf{r}}_i=v_0\left[ \begin{array}{c}
	\cos \theta _i\\
	\sin \theta _i\\
\end{array} \right]\\
	\dot{\theta}_i=\lambda \sum_{j\ne i}^N{A_{ij}\frac{\theta _j-\theta _i}{\left| \mathbf{r}_j-\mathbf{r}_i \right|^2}}\\
\end{cases}
$$


$$
\begin{aligned}
	A_{ij}&=\begin{cases}
	1,&		\phi _{ij}\in \left( \theta _i-\alpha ,\theta _i+\alpha \right)\\
	0,&		\mathrm{otherwise}\\
\end{cases}\\
	\phi _{ij}&=\mathrm{arc}\tan \frac{y_i-y_j}{x_i-x_j}\\
	\mathbf{r}_i&=\left[ \begin{array}{c}
	x_i\\
	y_i\\
\end{array} \right]\\
\end{aligned}
$$
