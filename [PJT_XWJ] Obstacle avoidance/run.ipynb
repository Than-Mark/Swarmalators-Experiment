$$
\begin{cases}
	\dot{\mathbf{r}}_i=v\left[ \begin{array}{c}
	\cos \theta _i\\
	\sin \theta _i\\
\end{array} \right]\\
	\dot{\theta}_i=\lambda \sum_{j\ne i}^N{\left( A_{ij}^{1}\frac{\phi _{ij}-\alpha -\theta _i}{\left| \mathbf{r}_i-\mathbf{r}_j \right|^2}+A_{ij}^{2}\frac{\phi _{ij}+\alpha -\theta _i}{\left| \mathbf{r}_i-\mathbf{r}_j \right|^2} \right)}\\
\end{cases}
$$

$$
\begin{aligned}
	A_{ij}^{1}&=\begin{cases}
	1,&		\theta _i\in \left( \phi _{ij}-\alpha ,\phi _{ij} \right)\\
	0,&		\theta _i\notin \left( \phi _{ij}-\alpha ,\phi _{ij} \right)\\
\end{cases}\\
	A_{ij}^{2}&=\begin{cases}
	1,&		\theta _i\in \left( \phi _{ij},\phi _{ij}+\alpha \right)\\
	0,&		\theta _i\notin \left( \phi _{ij},\phi _{ij}+\alpha \right)\\
\end{cases}\\
	\phi _{ij}&=\mathrm{arc}\tan \frac{y_i-y_j}{x_i-x_j}\\
	\mathbf{r}_i&=\left[ \begin{array}{c}
	x_i\\
	y_i\\
\end{array} \right]\\
\end{aligned}
$$
